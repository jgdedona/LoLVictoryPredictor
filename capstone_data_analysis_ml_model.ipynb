## Random Forest Classifier Model

A supervised random forest classification algorithm was used to implement the machine learning solution. Random forest classifiers are ensemble algorithms, meaning that their predictions are a cumulation of many subordinate decision tree models. The final prediction is derived from the accumulation of the subordinate models predictions.

This machine learning solution utilizes data collected directly from Riot Games, the company that developed and maintains League of Legends, to perform continued and up-to-date analysis on match characteristics. Specifically, this model uses a variety of features describing each match, such as gold earned, experience earned, objectives completed, and champions killed, as input in a random forest classification model to determine which team will win the match. This classification model will be of primary use to professional sportscasters, esports oddsmakers, and League of Legends enthusiasts because it will facilitate the identification of correlations between match objectives and match victory. This data can be used by the aforementioned groups to enhance informed sportscasting, set esports gambling odds, and stay up to date on current League of Legends statistical trends, respectively.

This model achieved approximately 97% accuracy in its victory predictions utilizing the average accuracy of 5 cross-validated training and test samples.

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

np.random.seed(92)

X = match_stats_df.drop("T1win", axis=1)
y = match_stats_df["T1win"]

clf = RandomForestClassifier(n_estimators=100)

score = np.mean(cross_val_score(clf, X, y, cv=5))
#score
```

```python
np.random.seed(92)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = RandomForestClassifier(n_estimators=100)

clf.fit(X_train, y_train);

#clf.score(X_test, y_test)
```

## Predictions

#### Enter statistics into the text fields below and the machine learning model will provide a real-time prediction of the probabilities of each team winning

```python
out = widgets.Output()
display(out)
```

```python
def on_value_change(T1gold,T1minions,T1dead,T1wardsK,T1wardsP,T1bar,T1champ,
                    T1drag,T1inhib,T1heralds,T1towers,T2gold,
                    T2minions,T2dead,T2wardsK,T2wardsP,T2bar,T2champ,
                    T2drag,T2inhib,T2heralds,T2towers):
    
    pred_dict = {'T1goldEarned':[T1gold], 'T1totalMinionsKilled':[T1minions],
     'T1totalTimeSpentDead':[T1dead], 'T1wardsKilled':[T1wardsK],
     'T1wardsPlaced':[T1wardsP], 'T2goldEarned':[T2gold],
     'T2totalMinionsKilled':[T2minions], 'T2totalTimeSpentDead':[T2dead],
     'T2wardsKilled':[T2wardsK], 'T2wardsPlaced':[T2wardsP],
     'T1baron':[T1bar], 'T1champion':[T1champ], 'T1dragon':[T2drag],
     'T1inhibitor':[T1inhib], 'T1riftHerald':[T1heralds],
     'T1tower':[T1towers], 'T2baron':[T2bar], 'T2champion':[T2champ],
     'T2dragon':[T2drag], 'T2inhibitor':[T2inhib],
     'T2riftHerald':[T2heralds], 'T2tower':[T2towers]}
    
    prediction_df = pd.DataFrame(data=pred_dict)
    prob = clf.predict_proba(prediction_df)
    pred = clf.predict(prediction_df)
    pred_string = 'Probability of Team One Victory: {T1}%\nProbability of Team Two Victory: {T2}%'

    with out:
        out.clear_output()
        print(pred_string.format(T1=prob[0][1]*100, T2=prob[0][0]*100))
```

```python
widgets.interactive(on_value_change,
            T1gold=T1goldEarned,
            T1minions=T1totalMinionsKilled,
            T1dead=T1totalTimeSpentDead,
            T1wardsK=T1wardsKilled,
            T1wardsP=T1wardsPlaced,
            T1bar=T1baron,
            T1champ=T1champion,
            T1drag=T1dragon,
            T1inhib=T1inhibitor,
            T1heralds=T1riftHerald,
            T1towers=T1tower,
            T2gold=T2goldEarned,
            T2minions=T2totalMinionsKilled,
            T2dead=T2totalTimeSpentDead,
            T2wardsK=T2wardsKilled,
            T2wardsP=T2wardsPlaced,
            T2bar=T2baron,
            T2champ=T2champion,
            T2drag=T2dragon,
            T2inhib=T2inhibitor,
            T2heralds=T2riftHerald,
            T2towers=T2tower)
```
